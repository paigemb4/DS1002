working w/ missing data

```python
import numpy as np
```

```python
df_miss = pd.DataFrame({
    'x':[2, np.nan, 1],
    'y':[np.nan, np.nan, 6]}
)
df_miss.dtypes
#nan means nothings there
```

```python
df_miss
```

```python
#.dropna() drops all rows w/ missing data from column
df_drop_all = df_miss.dropna()
df_drop_all
```

```python
df_drop_x = df_miss.dropna(subset=['x'])
df_drop_x
#subset takes a list of column names to specify which columns should test for missing values.
```

```python
ds_filled = df_miss.fillna(df_miss.median())
#.fillna replaces missing values to what you choose to set them to
ds_filled
```
