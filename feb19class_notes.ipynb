```python
import numpy as np
```

```python
#how to set two arrays = to each other:

a= np.array([[1,2,3],[1,2,3]])
b= np.array([[1,2,3],[1,2,3]])

print(a==b)

f = a[a==b]

print(f)
```

```python
#to show where arrays are not equal to each other:
c=a[a!=b] #all entries of a, where a is in a position that is not equal to b
print(c) #an array of nothing- length of 0
print(c.shape) #touple with a length 0
```
