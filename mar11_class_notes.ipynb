```python
#making data frames

df = pd.DataFrame({
    'x': [0, 2, 1, 5],
    'y': [1, 1, 0, 0],
    'z': [True, False, False, False]
})
#dictionaries have keys and values
#values come after colon
#keys are in quotes in front
df
```

**2nd way to make a data frame:**

Passing the three required pieces:

- columns as list
- index as list
- data as list of lists

```python
df2 = pd.DataFrame(
    columns=['x','y'],
    index=['row1','row2','row3'],
    data=[[9,3],[1,2],[4,6]])
df2
```

```python
my_data = [
    ('a', 1, True),
    ('b', 2, False)
]
df3 = pd.DataFrame(my_data, columns=['f1', 'f2', 'f3']) #turns it into a data frame

df3
```

```python
#can name indexes
df3.index.name = 'obs_id'
df3
```

- deep copy: makes copy of whole object
- shallow copy: copies the pointer (something that points to the information) can be changed

```python
df = pd.DataFrame({'x':[0,2,1,5], 'y':[1,1,0,0], 'z':[True,False,False,False]})

df_deep    = df.copy()  # deep copy; changes to df will not pass through
df_shallow = df         # shallow copy; changes to df will pass through

print('--df')
print(df)

# update values in df.x
df.x = 1

print('--Updated df')
print(df)
print('--df_shallow')
print(df_shallow)
print('--df_deep')
print(df_deep)

# rebuild df
df = pd.DataFrame({'x':[0,2,1,5], 'y':[1,1,0,0], 'z':[True,False,False,False]})
```
