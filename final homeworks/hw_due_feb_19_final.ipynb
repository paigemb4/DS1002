```python
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt
import io
from io import BytesIO
```

2.

```python
def red_filter(image):
  image_np = np.array(image)
  red_filtered_image_np = image_np.copy()
  red_filtered_image_np[:,:,1] = 0
  red_filtered_image_np[:,:,2]= 0
  red_filtered_image = Image.fromarray(red_filtered_image_np)
  return red_filtered_image

def green_filter(image):
    image_np = np.array(image)
    green_filtered_image_np = image_np.copy()
    green_filtered_image_np[:, :, 0] = 0
    green_filtered_image_np[:, :, 2] = 0
    green_filtered_image = Image.fromarray(green_filtered_image_np)
    return green_filtered_image

def blue_filter(image):
    image_np = np.array(image)
    blue_filtered_image_np = image_np.copy()
    blue_filtered_image_np[:, :, 0] = 0
    blue_filtered_image_np[:, :, 1] = 0
    blue_filtered_image = Image.fromarray(blue_filtered_image_np)
    return blue_filtered_image
```

function for all other filters:

```python
def red_green_filter(image):
    image_np = np.array(image)
    red_green_filtered_image_np = image_np.copy()
    red_green_filtered_image_np[:, :, 2] = 0
    red_green_filtered_image = Image.fromarray(red_green_filtered_image_np)
    return red_green_filtered_image

def red_blue_filter(image):
    image_np = np.array(image)
    red_blue_filtered_image_np = image_np.copy()
    red_blue_filtered_image_np[:, :, 1] = 0
    red_blue_filtered_image = Image.fromarray(red_blue_filtered_image_np)
    return red_blue_filtered_image

def green_blue_filter(image):
    image_np = np.array(image)
    green_blue_filtered_image_np = image_np.copy()
    green_blue_filtered_image_np[:, :, 0] = 0
    green_blue_filtered_image = Image.fromarray(green_blue_filtered_image_np)
    return green_blue_filtered_image
```

3. main script

1st picture:

```python
def url_to_image_array(image_url):
    response = requests.get(image_url)
    image_array = np.array(Image.open(BytesIO(response.content)))
    return image_array

url = input("Enter the URL of the image: ")
image_array = url_to_image_array(url)

new_picture = Image.fromarray(image_array)

red_filtered_image = red_filter(new_picture)
blue_filtered_image = blue_filter(new_picture)
green_filtered_image = green_filter(new_picture)
green_blue_filtered_image = green_blue_filter(new_picture)
red_blue_filtered_image = red_blue_filter(new_picture)
red_green_filtered_image = red_green_filter(new_picture)

plt.figure(figsize=(15, 5))

plt.subplot(2, 4, 1)
plt.imshow(red_filtered_image)
plt.title('Red Filtered')

plt.subplot(2, 4, 2)
plt.imshow(blue_filtered_image)
plt.title('Blue Filtered')

plt.subplot(2, 4, 3)
plt.imshow(green_filtered_image)
plt.title('Green Filtered')

plt.subplot(2, 4, 5)
plt.imshow(green_blue_filtered_image)
plt.title('Green Blue Filtered')

plt.subplot(2, 4, 6)
plt.imshow(red_blue_filtered_image)
plt.title('Red Blue Filtered')

plt.subplot(2, 4, 7)
plt.imshow(red_green_filtered_image)
plt.title('Red Green Filtered')

plt.tight_layout()
plt.show()

original_image_array = np.array(new_picture)
combined_filtered_image_array = np.stack((red_filtered_image, green_filtered_image, blue_filtered_image), axis=-1)

identical = np.array_equal(original_image_array, combined_filtered_image_array)

if identical:
    print("The original image is identical to the combined filtered image.")
else:
    print("The original image is not identical to the combined filtered image.")
```
